```python
# Split to functions, fix variable names
def read_words(filename):
    words = []
    with open(filename, 'r') as f:
        for line in f:
            words.extend(line.split())
    return words


def count_words(words):
    counter = {}
    for word in words:
        if word in counter:
            counter[word] += 1
        else:
            counter[word] = 1
    return counter


def most_common_words(counter):
    sorted_words = sorted(counter.items(), key=lambda kv: kv[1], reverse=True)
    return sorted_words[:20]


for kv in most_common_words(count_words(read_words('input.txt'))):
    print(kv)
```

```python
# Use collections.Counter
from collections import Counter

def read_words(filename):
    words = []
    with open(filename, 'r') as f:
        for line in f:
            words.extend(line.split())
    return words


def count_words(words):
    return Counter(words)


def most_common_words(counter):
    return list(counter.most_common(20))


for kv in most_common_words(count_words(read_words('input.txt'))):
    print(kv)
```

```python
# Fix bug with uppercase
from collections import Counter

def read_words(filename):
    words = []
    with open(filename, 'r') as f:
        for line in f:
            words.extend(line.split())
    return words


def count_words(words):
    return Counter([word.lower() for word in words])


def most_common_words(counter):
    return list(counter.most_common(20))


for kv in most_common_words(count_words(read_words('input.txt'))):
    print(kv)
```
